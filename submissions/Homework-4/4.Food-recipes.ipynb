```python
# Unique values for the ingredients

from pandas import Series
ingredients = Series(food.values.ravel()).unique()
```

```python
enc = preprocessing.LabelEncoder()
label = pd.Series(food['Vietnamese'].values.ravel()).unique()
enc.fit(label)
list(enc.classes_)
features_1 = enc.transform(food['Vietnamese'])

enc = preprocessing.LabelEncoder()
label_2 = pd.Series(food['vinegar'].values.ravel()).unique()
enc.fit(label_2)
list(enc.classes_)
features_2 = enc.transform(food['vinegar'])


label_4 = pd.Series(food['mint'].values.ravel()).unique()
enc.fit(label_4)
list(enc.classes_)
features_4 = enc.transform(food['mint'])
```

```python
plt.scatter(features_2, features_1)
```

We could also use SVD to transform this dataset into a different dimension in order to give us a better understanding.

```python
enc = preprocessing.LabelEncoder()
label_3 = pd.Series(food['cilantro'].values.ravel()).unique()
enc.fit(label_3)
list(enc.classes_)
features_3 = enc.transform(food['cilantro'])
```

```python
plt.scatter(features_3, features_1)
```

```python
# Gathering the columns of latitude & longitude
feature = np.array([features_1,features_2,features_3,features_4])
# Normalize the training dataset
feature = preprocessing.StandardScaler().fit_transform(feature)
```

Trying to fit into clustering

### Dendogram with Regions and Frequency of Ingredients

```python
kmeans = KMeans(init='k-means++', n_clusters=3, n_init=10)
kmeans.fit_predict(feature)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_
```

```python
# Plot the scatter graph
fig, ax = plt.subplots(1,1,figsize=(10,10))
# I have 10 clusters, so I need 10 colors. Each colormap can produce 10 colors
# so I am mixing two colormaps that I tested that work well together
cmap = sns.color_palette("bright", n_colors=4)
cmap.extend(sns.hls_palette(n_colors=6, h=0.7, l=0.4, s=0.8))
```
